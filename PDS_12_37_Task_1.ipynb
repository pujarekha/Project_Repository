```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Reading the CSV file containing data on the impact of environmental factors on plant growth
data = pd.read_csv("/content/Impact of Environmental Factors/Tausif_Kabir_Impact_of_Environmental_Factors_on_Plant_Growth.csv")
# Displaying the first five rows of the dataset
data.head()
```

The data shows that plant growth is positively influenced by both temperature and light intensity. Higher temperatures and greater light intensity, as observed on January 4, 2024, resulted in the highest plant growth (16 cm). Conversely, lower temperature and light intensity on January 3, 2024, led to reduced growth (11 cm). This highlights the combined impact of these environmental factors on plant development, with optimal conditions promoting better growth.

```python
# Checking for missing values in each column of the dataset and displaying the count
print(data.isnull().sum())
```

The result shows there are no missing values in any column of the dataset, meaning all data for date, temperature, light intensity, and plant growth is complete and ready for analysis without requiring cleaning or imputation.

```python
# Extracting the data types of each column in the dataset and storing them in a variable
data_types=data.dtypes
# Displaying the data types of the columns
data_types
```

The result shows the data types of each column in the dataset: Date is stored as an object (likely a string or datetime), Temperature (°C) is a float (decimal values), and both Light_Intensity (lux) and Plant_Growth (cm) are integers. This indicates the data types are appropriate for their respective columns, allowing for accurate analysis and calculations.

```python
# Generating summary statistics for the numerical columns in the dataset
summary_stats = data.describe()
# Displaying the summary statistics
summary_stats
```

The result provide insights into the dataset's distribution. For Temperature (°C), the average is 22.97°C, ranging from 21.2°C to 25.1°C, with low variability (std = 0.95). Light Intensity (lux) averages 330 lux, spanning 270 to 410 lux. Plant Growth (cm) averages 13.63 cm, ranging from 9 cm to 18 cm. The quartiles show the middle 50% of data, indicating most plant growth occurs between 12 and 15.75 cm under typical temperature and light conditions. This highlights consistent environmental factors and plant growth patterns.

```python
# Creating a figure with a specified size to plot two subplots side by side
fig_h = plt.figure(figsize=(33,9))

# Adding the first subplot to visualize the relationship between plant growth and temperature
fig1 = fig_h.add_subplot(1,2,1)
sns.histplot(data, y= "Plant_Growth (cm)", x="Temperature (°C)",hue="Plant_Growth (cm)", element="poly")
fig1.set_title("Plant Growth Distribution Under Varying Temperature Conditions")
fig1.set_xlabel("Temperature (°C)")  # Labeling the x-axis
fig1.set_ylabel("Plant Growth (cm)") # Labeling the y-axis
legend = fig1.get_legend()           # Accessing the legend of the first subplot
legend.set_bbox_to_anchor((1, .6))   # Adjusting the position of the legend

# Adding the second subplot to visualize the relationship between plant growth and light intensity
fig2 = fig_h.add_subplot(1,2,2)
sns.histplot(data, y = "Plant_Growth (cm)", x= "Light_Intensity (lux)", hue="Plant_Growth (cm)", element="poly")
fig2.set_title("Plant Growth Distribution Under Varying Temperature Conditions")
fig2.set_xlabel("Temperature (°C)")   # Labeling the x-axis
fig2.set_ylabel("Plant Growth (cm)")  # Labeling the y-axis
legend2 = fig2.get_legend()           # Accessing the legend of the first subplot
legend2.set_bbox_to_anchor((1, .6))   # Adjusting the position of the legend

# Adjusting the space between the two subplots to reduce overlap
fig_h.subplots_adjust(wspace=0.1)
# Displaying the figure
plt.show()
```

The figure display the distribution of plant growth (in cm) under varying temperature (left) and light intensity (right) conditions. Darker shades represent higher plant growth. Both graphs show a clear upward trend, indicating that higher temperatures and greater light intensity are associated with increased plant growth. This supports the idea that both factors positively influence plant development within the observed ranges.

```python
# Creating a figure with a specified size to plot two boxplots side by side
fig = plt.figure(figsize=(33,9))

# Adding the first subplot to visualize the distribution of plant growth across different temperature conditions
fig1 = fig.add_subplot(1,2,1)
sns.boxplot(x= "Temperature (°C)", y = "Plant_Growth (cm)", data= data, ax=fig1)
fig1.set_title("Plant Growth Distribution Under Varying Temperature Conditions")  # Setting the title for the first boxplot
fig1.set_xlabel("Temperature (°C)")    # Labeling the x-axis
fig1.set_ylabel("Plant Growth (cm)")   # Labeling the y-axis

# Adding the second subplot to visualize the distribution of plant growth across different light intensity conditions
fig2 = fig.add_subplot(1,2,2)
sns.boxplot(y = "Plant_Growth (cm)", x= "Light_Intensity (lux)", data= data, ax=fig2)
fig2.set_title("Plant Growth Distribution Under Varying Light Intensity Conditions")   # Setting the title for the second boxplot
fig2.set_xlabel("Light Intensity (lux)")  # Labeling the x-axis
fig2.set_ylabel("Plant Growth (cm)")      # Labeling the y-axis

# Adjusting the space between the two subplots to avoid overlapping
fig.subplots_adjust(wspace=0.1)
# Displaying the figure
plt.show()
```

The boxplots illustrate the distribution of plant growth under varying temperature (left) and light intensity (right) conditions. The graphs show an upward trend in plant growth as temperature and light intensity increase. The consistent medians and narrow interquartile ranges at higher levels suggest that both factors positively influence growth, with less variability. Outliers in the data indicate occasional deviations from the trend.

```python
# Creating a figure with a specified size to plot two scatter plots side by side
fig_s = plt.figure(figsize=(33,9))

# Adding the first subplot to visualize the relationship between plant growth and temperature
fig1 = fig_s.add_subplot(1,2,1)
sns.scatterplot(x= "Temperature (°C)", y = "Plant_Growth (cm)", data= data, color = "red")
fig1.set_title("Plant Growth vs Temperature")  # Setting the title for the first scatter plot
fig1.set_xlabel("Temperature (°C)") # Labeling the x-axis
fig1.set_ylabel("Plant Growth (cm)") # Labeling the y-axis

# Adding the second subplot to visualize the distribution of plant growth across different light intensity conditions
fig2 = fig_s.add_subplot(1,2,2)
sns.scatterplot(y = "Plant_Growth (cm)", x= "Light_Intensity (lux)", data= data, color="blue")
fig2.set_title("Plant Growth vs Light Intensity")  # Setting the title for the second scatter plot
fig2.set_xlabel("Light Intensity (lux)")   # Labeling the x-axis
fig2.set_ylabel("Plant Growth (cm)")       # Labeling the y-axis

# Adjusting the space between the two subplots to avoid overlapping
fig_s.subplots_adjust(wspace=0.1)
# Displaying the figure
plt.show()
```

The scatter plots show a clear positive relationship between both temperature and light intensity, and plant growth. As temperature and light intensity increase, so does the plant growth. This visual representation supports the findings of the correlation matrix, suggesting that both temperature and light intensity are likely contributing factors to plant growth. However, it's important to remember that correlation does not equal causation, and further research is needed to establish the exact causal relationships between these variables.

```python
# Calculating the correlation matrix for selected numerical columns
correlation_matrix = data[["Temperature (°C)", "Light_Intensity (lux)", "Plant_Growth (cm)"]].corr()
# Displaying the correlation matrix
correlation_matrix

# Creating a heatmap to visualize the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='prism_r', vmin=-1, vmax=1)  # Adding annotations and a color map
plt.title('Correlation Matrix')   # Setting the title for the heatmap
plt.show()
```

The correlation matrix indicates a strong positive relationship between temperature, light intensity, and plant growth. This suggests that higher temperatures and increased light intensity are likely to promote greater plant growth. However, it's important to remember that correlation does not equal causation, and further research is needed to determine the specific causal relationships between these variables.

```python
# Finding the column that has the strongest correlation with 'Plant_Growth (cm)'
# We drop 'Plant_Growth (cm)' from the correlation matrix, then find the column with the highest absolute correlation value
strongest_corr = correlation_matrix['Plant_Growth (cm)'].drop('Plant_Growth (cm)').abs().idxmax()
# Printing the column that has the strongest correlation with plant growth
print(f"The Strongest correlation with plant growth is: {strongest_corr}")
```

```python
# Creating a figure with a specified size to plot two regression plots side by side
fig_l = plt.figure(figsize=(25,9))

# Adding the first subplot to visualize the linear regression between temperature and plant growth
fig1 = fig_l.add_subplot(1,2,1)
sns.regplot(x="Temperature (°C)", y='Plant_Growth (cm)', data=data, ax=fig1)  # Fitting a linear regression line
fig1.set_title('Linear Regression: Temperature vs. Plant Growth')  # Setting the title for the first plot

# Adding the second subplot to visualize the linear regression between light intensity and plant growth
fig2 = fig_l.add_subplot(1,2,2)
sns.regplot(x="Light_Intensity (lux)", y='Plant_Growth (cm)', data=data, ax=fig2)  # Fitting a linear regression line
fig2.set_title('Linear Regression: Light Intensity vs. Plant Growth')   # Setting the title for the second plot

# Adjusting the space between the two subplots to reduce overlap
fig_l.subplots_adjust(wspace=0.1)
# Displaying the figure
plt.show()
```

The graphs show linear regression analyses of plant growth against temperature and light intensity. In the first graph (left), plant growth positively correlates with temperature, suggesting that higher temperatures (within the observed range) promote plant growth. In the second graph (right), plant growth also shows a strong positive correlation with light intensity, indicating that increased light intensity contributes significantly to plant growth. Both factors appear to have a direct and measurable impact on the plant's development, with light intensity showing a particularly strong effect.

```python
# Converting the 'Date' column to datetime format for better date-related operations
data['Date'] = pd.to_datetime(data['Date'])

# Setting the 'Date' column as the index of the DataFrame to make date-based indexing easier
data.set_index('Date', inplace=True)
```

```python
# Creating a line plot to visualize plant growth over time
plt.figure(figsize=(12, 6))   # Setting the size of the plot
plt.plot(data.index, data['Plant_Growth (cm)'])  # Plotting the plant growth over time using the 'Date' index
plt.title('Plant Growth Over Time')   # Setting the title of the plot
plt.xlabel('Date')    # Labeling the x-axis as 'Date'
plt.ylabel('Plant Growth (cm)')   # Labeling the y-axis as 'Plant Growth (cm)'
plt.show()   # Displaying the plot
```

This line graph depicts plant growth over time, measured in centimeters, from January 1, 2024, to February 22, 2024. The data shows fluctuations in growth, with plant height varying between approximately 10 cm and 18 cm. The trend suggests that growth was inconsistent during this period, potentially influenced by environmental factors, care routines, or seasonal changes.

```python
# Creating a line plot to visualize both temperature and light intensity over time
plt.figure(figsize=(12, 6))  # Setting the size of the plot

# Plotting the 'Temperature (°C)' over time with a label for the legend
plt.plot(data.index, data["Temperature (°C)"], label='Temperature')

# Plotting the 'Light_Intensity (lux)' over time with a label for the legend
plt.plot(data.index, data['Light_Intensity (lux)'], label='Light Intensity')

# Adding a title to the plot
plt.title('Temperature and Light Intensity Over Time')
# Labeling the x-axis as 'Date'
plt.xlabel('Date')
# Labeling the y-axis as 'Values' since both temperature and light intensity are being plotted
plt.ylabel('Values')
# Displaying the legend to differentiate the two lines (temperature and light intensity)
plt.legend()
# Displaying the plot
plt.show()
```

This graph compares temperature and light intensity over time, from January 1, 2024, to February 22, 2024. Light intensity (orange line) shows significant fluctuations, ranging between 300 and 450 units, indicating variability in lighting conditions. In contrast, temperature (blue line) remains relatively stable throughout, with values staying consistently low. The stable temperature and fluctuating light intensity suggest that lighting may have played a more prominent role in influencing plant growth during this period.
